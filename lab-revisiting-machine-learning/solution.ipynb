## Lab | Handling Data Imbalance in Classification Models

```python
donors = pd.read_csv('learningsetclean.csv', low_memory=False)
```

```python
donors.head()
```

### Modeling: Train/Test Split

```python
from sklearn.model_selection import train_test_split

# Get X and Y
X = donors.drop(['TARGET_B','TARGET_D'],axis=1)
y = donors['TARGET_B']

# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Numericals split
X_train_num = X_train.select_dtypes(np.number)
X_test_num = X_test.select_dtypes(np.number)
display(X_train_num.head())

# Categoricals split
X_train_cat = X_train.select_dtypes(object)
X_test_cat = X_test.select_dtypes(object)
display(X_train_cat.head())
```

### Scale the features

```python
from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler().fit(X_train_num)
X_train_num_standardized = transformer.transform(X_train_num)
X_train_num_standardized = pd.DataFrame(X_train_num_standardized, columns=X_train_num.columns, index = X_train_num.index)
display(X_train_num_standardized.head())

X_test_num_standardized = transformer.transform(X_test_num)
X_test_num_standardized = pd.DataFrame(X_test_num_standardized, columns=X_test_num.columns, index = X_test_num.index)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first', handle_unknown='error').fit(X_train_cat)
hot_train_encoded = encoder.transform(X_train_cat).toarray()
hot_test_encoded = encoder.transform(X_test_cat).toarray()

cols = encoder.get_feature_names_out(input_features=X_train_cat.columns)
hot_train_cat = pd.DataFrame(hot_train_encoded, columns=cols, index = X_train_cat.index)
hot_test_cat = pd.DataFrame(hot_test_encoded, columns=cols, index = X_test_cat.index)
```

```python
display(hot_train_cat.head())
```

```python
X_train = pd.concat((X_train_num_standardized,hot_train_cat),axis=1)
X_test = pd.concat((X_test_num_standardized,hot_test_cat),axis=1)

display(X_train.head())
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import plot_precision_recall_curve
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
# Model

LR = LogisticRegression(solver='saga',multi_class='multinomial',
                        max_iter=1000).fit(X_train, y_train) #fit model on training data
```

```python
# Model Validation

predictions = LR.predict(X_test)
LR.score(X_test, y_test)
```

```python
# Confusion Matrix

confusion_matrix(y_test, predictions)
```

```python
print(y_test.value_counts())
```

### Handle Imbalance (downsample)

```python
trainset = pd.concat([X_train, y_train], axis=1)
```

```python
trainset['TARGET_B'].value_counts()
```

```python
from sklearn.utils import resample

category_0 = trainset[trainset['TARGET_B'] == 0]
category_1 = trainset[trainset['TARGET_B'] == 1]
```

```python
#Downsample
category_0_undersampled = resample(category_0, 
                                   replace=False, 
                                   n_samples = len(category_1))
```

```python
print(category_0.shape)
print(category_1.shape)
```

```python
print(category_0_undersampled.shape)
```

```python
data_downsampled = pd.concat([category_0_undersampled, category_1], axis=0)
```

#### Fit model with new downsampled data

```python
# Get X and Y
X_train_down = data_downsampled.drop(['TARGET_B'],axis=1)
y_train_down = data_downsampled['TARGET_B']
```

```python
LR_downsample = LogisticRegression(solver='saga',multi_class='multinomial',
                        max_iter=1000).fit(X_train_down, y_train_down) 

LR_downsample.score(X_test, y_test)
```

```python
down_pred = LR_downsample.predict(X_test)
confusion_matrix(y_test, down_pred)
```

### Handle Imbalance (Upsampling)

```python
category_1_oversampled = resample(category_1, 
                                  replace=True, 
                                  n_samples = len(category_0))
```

```python
print(category_0.shape)
print(category_1.shape)
```

```python
print(category_1_oversampled.shape)
```

```python
data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
```

#### Fit model with new oversampled data

```python
# Get X and Y
X_train_up = data_upsampled.drop(['TARGET_B'],axis=1)
y_train_up = data_upsampled['TARGET_B']
```

```python
LR_oversample = LogisticRegression(solver='saga',multi_class='multinomial',
                        max_iter=1000).fit(X_train_up, y_train_up) 

LR_oversample.score(X_test, y_test)
```

```python
over_pred = LR_oversample.predict(X_test)
confusion_matrix(y_test, over_pred)
```
